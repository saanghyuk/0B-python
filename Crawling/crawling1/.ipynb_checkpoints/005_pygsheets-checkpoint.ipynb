### 구글 스프레드시트에 저장


- oauth 개념 : user들이 생성한 데이터는 페이스북이 가지고 있음. 그런데, 이 데이터는 페이스북과 유저가 동시에 저작권 가지고 있음. 이 데이터 가져오려면 페이스북과 유저에게 둘다 허락을 받아야함. 그 허락을 받는 것이 `oauth`. 
    - 1. Facebook에 등록
    - 2. Facebook에서 나한테 request token을 발행
    - 3. 그럼 이제 유저한테 허락을 받아야 함. user1이 내 사이트에 들어오면 `너 정보 페이스북에서 가져올 수 있도록 인증 해줘` 요청. `허용` 버튼. 허용하기를 누르면, 나한테 access token이 생긴다. 
    - 4. 이제 나는 request token과 user1의 access token으로 페이스북에 다시 요청. 
    - 5. 페이스북에서 내 토큰들을 보고 데이터를 보내준다. 
        - 참고로 페이스북에서 친구리스트까지 다 가져갈 수 있음. 
        - `허락하기`는 조심해서 눌러야 함.

```python
import pygsheets
gc = pygsheets.authorize(client_secret="client_secret.json")
```

```python
# 내부에서 request, response 알아서 해주는 객체

gc
```

```python
# 스프레드시트 파일 열기
s_file = gc.open("zigbang")
```

```python
s_file
```

```python
# 붓꽃 데이터 로드
import seaborn as sns
import pandas as pd
iris_df = sns.load_dataset("iris")
iris_df.tail(2)
```

```python
# iris 시트 만들기
iris_sheet = s_file.add_worksheet("iris")
```

```python
# iris 시트에 데이터 프레임 저장
iris_sheet.set_dataframe(iris_df, "A1")
```

```python
s_file[1]
```

```python
# 시트 삭제
s_file.del_worksheet(s_file[0])
```

```python
# 직방 데이터를 크롤링 해서 저장
```

```python
addrs = ["연남동", "망원동", "합정동", "성수동", "개포동"]
for addr in addrs:
    df = zb.crawling_apt(addr)
    sheet = s_file.add_worksheet(addr)
    sheet.set_dataframe(df, "A1", copy_index=False)
    print(addr)
```

```python
# 데이터 가져오기
# sheet title "성수동" case
sheet = s_file.worksheet_by_title("성수동")
load_df = sheet.get_as_df()
len(load_df)
```
