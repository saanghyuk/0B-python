## 네이버 연관 검색어 가져오기

```python
# 1. 웹페이지 분석 : URL 찾기

query = "삼성전자"
url = "https://search.naver.com/search.naver?ie=UTF-8&query={}&sm=chr_hty".format(query)
print(url)
```

```python
# 2. request -> response : HTML(str)
import requests

response = requests.get(url)
response
```

```python
# 3. HTML(str) -> BeautifulSoup Obj (css-selector) -> DF

from bs4 import BeautifulSoup
import pandas as pd

dom = BeautifulSoup(response.content, "html.parser")
type(dom)
```

```python
# dom
# select : 여러개의 element 객체를 리스트로 가져옴
# select_one : 하나의 element 객체를 가져옴
```

```python
elements = dom.select(".lst_related_srch > li")
len(elements)
```

```python
keywords = []
for element in elements:
    keywords.append(element.text.strip())
    
print(keywords)
```

```python
keywords = [element.text.strip() for element in elements]
keywords
```

```python
# 함수로 만들기
def naver_relational_keywords(query):
    url = "https://search.naver.com/search.naver?ie=UTF-8&query={}&sm=chr_hty".format(query)
    response = requests.get(url)
    dom = BeautifulSoup(response.content, "html.parser")
    elements = dom.select(".lst_related_srch > li")
    
    return [element.text.strip() for element in elements]
```

```python
query = "코로나"
result = naver_relational_keywords(query)
print(result)
```

```python
# 1. URL media.daum.net 에서 news.daum.net 으로 변경 되었습니다.
url = "https://news.daum.net/"
```

```python
# 2. HTML(str)
response = requests.get(url)
response
```

```python
# 3. BS
dom = BeautifulSoup(response.content, "html.parser")
```

```python
elements = dom.select(".list_headline > li")
len(elements)
```

```python
elements[0]
```

```python
datas = []

for element in elements:
    datas.append({
        "title": element.select_one('.tit_g').text.strip().replace("\n", ""),
        "link": element.select_one('a').get('href'),
    })
    
articles_df = pd.DataFrame(datas)
articles_df.tail(2)
```

```python
# content 데이터 추가
```

```python
link = articles_df.loc[0, "link"]
```

```python
response = requests.get(link)
dom = BeautifulSoup(response.content, "html.parser")
```

```python
content = dom.select_one("#harmonyContainer").text.strip().replace("\n", "")
```

```python
# summary_view 가 있는 기사

response = requests.get("https://news.v.daum.net/v/20210423124921366")
dom = BeautifulSoup(response.content, "html.parser")

content = dom.select_one(".summary_view").text.strip().replace("\n", "")
content
```

```python
# 함수 만들기
```

```python
def get_content(link):
    response = requests.get(link)
    dom = BeautifulSoup(response.content, "html.parser")
    return dom.select_one("#harmonyContainer").text.strip().replace("\n", "")
```

```python
articles_df["contents"] = articles_df["link"].apply(get_content)
articles_df.tail(2)
```
