```python
import scipy as sp
import scipy.stats
```

```python
rv = sp.stats.norm(loc=1, scale=2)
```

```python
rv.pdf
```

```python
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('PDF')
```

```python
rv_1 = sp.stats.norm(loc=0, scale=0.1)
xx = np.linspace(-1, 1, 100)
pdf = rv_1.pdf(xx)
plt.plot(xx, pdf)
plt.title('PDF')
```

```python
N=50000
rvs = sp.stats.uniform().rvs(N)
for i in range(9):
    rvs_temp = sp.stats.uniform().rvs(N)
    rvs = rvs + rvs_temp
```

```python
len(rvs)
```

```python
sns.distplot(rvs, kde=False)
```

Binomial

```python
N=10
mu = 0.65
rv = sp.stats.binom(N, mu)
```

```python
xx = np.arange(N+1)
plt.bar(xx, rv.pmf(xx), align='center')
rv.pmf(xx)
```
