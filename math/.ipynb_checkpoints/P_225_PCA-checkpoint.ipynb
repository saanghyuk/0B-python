### 이미지 PCA

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
faces_all = fetch_olivetti_faces()
```

```python
K = 20 # 20번 사진 인물만 선택

#이미지 함수로 쓸때는 NxN꼴로 줌
faces = faces_all.images[faces_all.target == K]
faces.shape
```

```python
faces.shape
```

```python
faces.shape
```

```python
N = 2
M = 5
```

```python
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(faces[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
```

- 지금 현재 데이터 샘플(64 X 64)이 10개가 있는 것.
- 사진 하나마다, 벡터 x1, x2, x3 ... x10 까지

```python
from sklearn.decomposition import PCA
```

```python
pca3 = PCA(n_components=2)
# 데이터 함수를 쓰면, 4096차원 벡터로 줌
X3 = faces_all.data[faces_all.target == K] 
X_low = pca3.fit_transform(X3)

# 그 변환한다음에 그거를 다시 원래 기저로 옮겨야 우리가 그림도 그리고 볼 수 있잖아. 
# w1, w2만 알면 우리는 언제든지 원래 차원으로 가져갈 수 있음. 
# 아래는 이 X32로 그림 그린 것
X32 = pca3.inverse_transform(X_low)
```

```python
X3.shape
```

```python
print(pca3.components_.shape)
pca3.components_
```

```python
faces[1]
```

```python
X32[1]
```

```python
# 2차원으로 차원 축소됨. 
# 이것을 가지고, inverse_transform에서 원래 4096차원이였으면 어디였을지 읽었을 지 알려주는 것. 
X_low, X32.shape
```

```python
# 바뀐 원점 좌표, 원래 기저로
pca3.mean_.shape
```

```python
# w1과 w2(2차원으로 축소)
pca3.components_.shape
```

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(X32[k].reshape(64, 64), cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle("PCA")
plt.tight_layout()
plt.show()
```

- 자세히 보면 원래사진이랑 조금 다르다. 
- Approximation 상태. 
- 4096개를 가지고 숫자 2개로 압축한거야. 
- 물론 w1, w2는 4096차원 벡터. 이 w1, w2를 안다는 가정 하에 숫자 2개만 알면 위 이미지가 나오는 것. 
- 무선통신 같은 것도 다 이런걸로 압축하는 것. 압축 된것을 보내고, w1, w2는 미리 공유 해 놓는 것. 
- 이 미리 w1, w2 공유해 놓은것을 코드북이라고 하는 것. 

```python
# pca3.mean_은 4096차원 나옴. 왜? 각각 벡터별로의 평균을 의미하니깐. 
# # 새로운 원점. 새로운 원점을 4096차원 상에서 표기한 것. 
face_mean = pca3.mean_.reshape(64, 64)

# w1도 4096차원(사실 원래 이미지 벡터는 아니야, 방향을 가리키는 벡터야. 억지로 그린 것. 이렇게 제일 많이 변했다는 것)
face_p1 = pca3.components_[0].reshape(64, 64)
# w2도 4096차원(사실 원래 이미지 벡터는 아니야, 방향을 가리키는 벡터야. 억지로 그린 것. 2번째로 많이 변화한 방향)
face_p2 = pca3.components_[1].reshape(64, 64)

plt.subplot(131)
plt.imshow(face_mean, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("Average Face")
plt.subplot(132)
plt.imshow(face_p1, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("PCA 1")
plt.subplot(133)
plt.imshow(face_p2, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("PCA 2")
plt.show()
```

- 이미지를 합성해 보자. 
- 평균 벡터에, w벡터를 *1, *2, 이렇게 계속 더해본 것.
- 원래 있던 이미지는 아님. 

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k - 5) if k < 5 else 1.5 * (k - 4)
        ax.imshow(face_mean + w * face_p1, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("PCA1 Proportion={}".format(w))
plt.suptitle("AVG FACE + W Photo")
plt.tight_layout()
plt.show()

```

- 고개 방향이 달라지네. 
- 고개 방향이 왜 달라질까?
- w1은 데이터가 변화한 방향을 나타내는 건데, 
- 이미지 마다 데이터가 다르잖아. "고개가 오른쪽 왼쪽으로 바뀌는 것이 제일 큰 차이라고 말하는 것"

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k - 5) if k < 5 else 1.5 * (k - 4)
        ax.imshow(face_mean + w * face_p2, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("w2 proportion={:.1f}".format(w))
plt.suptitle("AVG FACE + W2")
plt.tight_layout()
plt.show()
```

- 이건 갈수록 점점 웃는다. 
- 즉, 웃는 것 웃지 않는 것 사이에 달라지는게 두번째 데이터의 차이라는 것. 

### 주식가격 PCA

```python
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
import datetime
```

```python

symbols = [
    "SPASTT01USM661N", # US: 미국
    "SPASTT01JPM661N", # JP: 일본
    "SPASTT01EZM661N", # EZ: 유럽
    "SPASTT01KRM661N", # KR: 한국
]

data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.DataReader(sym, data_source='fred', 
                               start=datetime.datetime(1998, 1, 1),
                               end=datetime.datetime(2017, 12, 31))[sym]
data.columns = ["US", "JP", "EZ", "KR"]
data = data / data.iloc[0] * 100

styles = ["b-.", "g--", "c:", "r-"]

data.plot(style=styles)
plt.title("20 years stock price")
plt.show()
```

- 앞의 차트는 시작시점의 주가가 100이 되도록 크기를 조정하였다. 이 데이터로부터 연간 주식수익률 데이터를 구하면 다음처럼 20×4 크기의 특징행렬을 구할 수 있다.
- 이 연간 주식수익률 데이터를 그래프로 나타내면 다음과 같다. 붓꽃의 경우보다 (2차원에서 20차원으로) 차원은 증가했지만 각 표본이 가지는 측정 데이터 벡터의 모양(비율)이 비슷하다는 것을 알 수 있다. 붓꽃이 측정 데이터의 값을 결정하는 “크기”라는 잠재변수를 가지고 있는 것처럼 각 나라의 수익률은 “경제적 요인”이라고 하는 공통된 잠재변수에 의해서 결정되기 때문이다.




```python
df = ((data.pct_change() + 1).resample("A").prod() - 1).T * 100
print(df.iloc[:, :5])
print(df.shape) # 4개 국가, 20개 features
```

```python
df.T.plot(style=styles)
plt.title("Profit Rate for 20 years")
plt.xticks(df.columns)
plt.show()
```

```python
pca2 = PCA(n_components=1)
X_low = pca2.fit_transform(df)
X_low.shape # (4, 1)
X_low
```

```python
# 20개 Feature를 가진 4개의 점들간의 평균 값
m = pca2.mean_
m.shape # 20, 
m
```

```python
# 이게 새로운 1차원 기저벡터
p1 = pca2.components_[0]
p1
```

이 두 가지 성분을 차트로 나타내보자. 평균주가로부터 주성분이 더해질수록 어떻게 주가의 형태가 바뀌는지 살펴보면 다음과 같다. 굵은 선으로 표시된 주가가 평균 주가이고 나머지 주가는 주성분이 점점 많이 추가됨에 따라 어떤 주가 형태로 변하는지를 표시한 것이다. 여기에서는 주성분의 모습을 잘 보기 위해 주성분을 과장하여 추가하였다.

```python
xrange = np.linspace(1998, 2017, 20, dtype=int)
for i in np.linspace(0, 100, 5):
    plt.plot(xrange, pca2.mean_ + p1 * i)
plt.plot(xrange, pca2.mean_ + p1 * 100, label="PCA + 100 times")
plt.plot(xrange, pca2.mean_, "ko-", lw=5, label="AVG PROFIT RATE")
plt.title("CHANGES OF THE PROFIT RATE WITH THE PCA")
plt.legend()
plt.show()
```

- 나라가 달라지면, 위처럼 변해간다는 말. 굵은 점이 평균
- 여기서는 굵은 평균선 빼고, 선 하나하나가 평균 + PCA추가 해 가면서 그린 선이야. 
- 즉, 이 변화의 방향이 2000년에는 국가가 변해갈수록 더 많이 떨어지는 방향, 2005년에는 국가가 변해갈수록 더 올라가는 방향으로 갔다는 것을 알 수 있지. 

이 주성분이 가지는 의미를 살펴보기 위해 각 나라의 주성분 비중을 구한다.



```python
# 1차원으로 차원축소한 벡터를 의미
X_low
```

```python
df_w = pd.DataFrame(X_low)
df_w.index = df.index
df_w.columns = ["주성분 비중"]
print(df_w)
```

- 한국은 주성분의 비중이 80이 넘고 나머지 국가들은 모두 주성분의 성분이 음수임을 알 수 있다. 또한 주성분을 이용하여 구한 근사 주가를 표시하면 다음과 같다.
- inverse_transform 해서 다시 구한 다음에 그린 것

```python
df_i = pd.DataFrame(pca2.inverse_transform(X_low))
df_i.index = df.index
df_i.columns = df.columns
df_i.iloc[:, -10:]
df_i.T.plot(style=styles)
plt.title("Yearly Profit Rate of Past 20 years, using PCA")
plt.xticks(df.columns)
plt.show()
```

이 결과로부터 우리가 구한 주성분은 ‘미국, 일본, 유럽’이라는 데이터로부터 ‘한국’이라는 데이터로 변화시키는 요인, 즉 중진국 요인이라 부를 수 있는 성분이라는 것을 명확하게 알 수 있다.
