```python
def heavyside_step(x):
    if isinstance(x, np.ndarray):
        return np.where(x >= 0, 1, 0)
    else:
        return 1.0 if x >= 0 else 0.0
```

```python
heavyside_step(-0.00001), heavyside_step(0), heavyside_step(0.001)
```

파이썬으로 함수 그리기

```python
def f(x):
    return x**3 - 3 * x**2 + x
```

```python
x = np.linspace(-1, 3, 9)
x
```

```python
y = f(x)
y
```

```python
plt.plot(x, y, 'ro-')
```

- 부드럽게 되있지가 않잖아.
- 안타깝게도 파이썬에서는 부드럽게 만드는 명령이 없음. 
- 부드럽게 만들고 싶으면 점을 많이 찍어줘야 함. 

```python
plt.plot(x, y, 'ro-')
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.show()
```

```python
x = np.linspace(-1, 3, 400)
y = f(x)
```

```python
plt.plot(x, y, 'ro-')
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.show()
```

부호함수 그려보기

```python
def sign_fun(x):
    if isinstance(x, np.ndarray):
        return np.where(x > 0, 1, np.where(x == 0, 0, -1))    
    else:
        return 1.0 if x > 0 else 0 if x == 0 else -1
```

```python
x = np.linspace(-1, 3, 10)
x
```

```python
y = sign_fun(x)
y
```

```python
plt.plot(x, y, 'ro-')
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.show()
```

단위계단함수 그려보기

```python
def heavyside_step(x):
    if isinstance(x, np.ndarray):
        return np.where(x >= 0, 1, 0)
    else:
        return 1.0 if x >= 0 else 0.0
```

```python
x = np.linspace(-1, 3, 10)
y = heavyside_step(x)
```

```python
plt.plot(x, y, 'ro-')
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-2, 2))
plt.show()
```
