```python
np.random.seed(0)

rv = sp.stats.norm()
M = 2000

plt.subplot(1, 2, 1)
N = 6
x = rv.rvs((N, M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1, 10)
plt.ylim(0, 800)
plt.title("(N = 1)")
plt.xlabel("Value")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
N = 100
x = rv.rvs((N, M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
# plt.xlim(-1, 10)
# plt.ylim(0, 800)
plt.title("(N = 100)")
plt.xlabel("Value")
plt.ylabel("Count")

plt.show()
```
