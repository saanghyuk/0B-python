### Basic Pre-Processing

이 절에서는 데이터를 본격적으로 분석하기 이전에 다음과 같은 패키지를 사용하여 기초적인 전처리(preprocessing)를 하는 방법을 설명한다.

- missingno 패키지: 결측 데이터 검색, 시각화
- sklearn.impute 패키지: 결측 데이터 대체(빠진 데이터 집어넣고 싶을 때)
- patsy 패키지: 데이터 선택, 변환, 추가, 스케일링(데이터 매우 많은 경우 선택, 변환 등)
- sklearn.preprocessing 패키지: 스케일링, 변환

- missingno 패키지¶
    - 현실에서 데이터를 수집하다보면 데이터의 일부를 얻지 못하거나 누락되는 결측(missing) 데이터가 생긴다. missingno 패키지는 pandas 데이터프레임에서 결측 데이터를 찾는 기능을 제공한다. 이 패키지를 사용하려면 데이터프레임에 결측 데이터가 **NaN(not a number)** 값으로 저장되어 있어야 한다.

    - 주의할 점은 *NaN값은 부동소수점 실수 자료형에만 있는 값이므로* 정수 자료를 데이터프레임에 넣을 때는 ***Int64Dtype*** 자료형을 명시해주어야 하고 시간 자료형을 넣을 때도 parse_dates 인수로 날짜시간형 파싱을 해주어야 ***datetime64[ns]*** 자료형이 되어 결측 데이터가 NaT(not a time) 값으로 표시된다.
    - 즉, 정수형이나 실수형 등에는 NaN이라는 것 자체가 없음. 다행이도 pandas에는 정수형이나 실수형 등에도 NaN을 쓸 수 있는 특별한 타입을 가지고 있음. 

```python
from io import StringIO

csv_data = StringIO("""
x1,x2,x3,x4,x5
1,0.1,"1",2019-01-01,A
2,,,2019-01-02,B
3,,"3",2019-01-03,C
,0.4,"4",2019-01-04,A
5,0.5,"5",2019-01-05,B
,,,2019-01-06,C
7,0.7,"7",,A
8,0.8,"8",2019-01-08,B
9,0.9,,2019-01-09,C
""")


# x1은 Int64Dtype자료형으로 읽고, 3번째 열은 parse_dates를 통해 날짜 타입으로 바꿔서 읽으라는 것을 명시. 
df = pd.read_csv(csv_data, dtype={"x1": pd.Int64Dtype()}, parse_dates=[3])
# df = pd.read_csv(csv_data) # 그냥 이렇게 해도 되네?
df
```

### **결측된 데이터를 처리하는 방법은 두 가지다.**

- 일반적으로는 남은 데이터 중에 대표값으로 나머지를 채움. 가장 일반적.

fit_transform 메서드를 사용하여 대체값이 채워신 데이터프레임을 생성한다.

strategy 인수를 선택하는 방법은 다음과 같다.

데이터가 실수 연속값인 경우에는 평균 또는 중앙값을 사용할 수 있다.<br>
***값의 분포가 대칭적이면 평균이 좋고 값의 분포가 심하게 비대칭인 경우에는 중앙값이 적당하다.***

### patsy 패키지

가장 중요한 패키지 중 하나, 전처리에서 할 수 있는 대부분의 것들을 할 수 있는 기능을 다 가지고 있다. 


이번에는 patsy 패키지를 사용하여 데이터프레임에서 원하는 데이터만 선택하거나 새로운 데이터를 조합 생성하는 방법을 살펴본다. 설명을 위해 patsy 패키지가 제공하는 demo_data() 함수로 다음과 같은 예제 데이터프레임을 만들자. demo_data() 함수는 x로 시작하는 변수에 대해 임의의 실수 데이터를 생성한다.

```python
from patsy import demo_data

df = pd.DataFrame(demo_data("x1", "x2", "x3", "x4", "x5"))
df
```

우리가 데이터 분석을 할 때, 보통은 데이터프레임 안에 넣어놓고 시작한다. 데이터 종류가 수백개 수천개 수만개씩 된다. 
보통은 일단 데이터프레임 하나에 다 때려박아 놓는다. <br>
집어넣다고 해서 다 쓰는 것은 아님. 분석할때는, 몇개씩 빼서 분석해보고 이렇게 함. <br>
***전체 다 들어있는 데이터프레임에서 일부만 선택을 하는 방법***을 일단 patsy가 지원을 해준다. 




데이터 분석을 시작할 때는 다양한 데이터를 하나의 데이터프레임에 넣고 시작하는 경우가 많다. patsy 패키지가 제공하는 dmatrx() 함수를 사용하면 데이터 프레임에 상수항을 추가하거나 원하는 데이터만 선택하거나 변형할 수 있다. dmatrx() 함수에 모형 정의 ***문자열 formula***와 원 데이터를 담은 데이터프레임 data을 입력하면 formula에서 지정한 대로 변환된 데이터 data_transformed를 출력한다.

`data_transformed = dmatrix(formula, data)`

formula는 데이터 열 이름 기반으로 구성된 문자열이다. 선택하고자 하는 데이터 열 이름을 +로 연결한 formula 문자열을 입려하면 자동으로 해당 데이터만 뽑아준다. 예를 들어 전체 데이터 중 x1만을 뽑고 싶으면 formula 문자열에 x1 + 0이라고 입력한다.

```python
from patsy import dmatrix 

#df라는 통에다 집어놓고, x1만 뽑고 싶은 경우
dmatrix("x1 + 0", data=df)
```

전체 데이터 중 x1, x2, x3를 뽑고 싶으면 formula 문자열에 x1 + x2 + x3 + 0이라고 입력한다.

```python
dmatrix("x1 + x2 + x3 + 0", data=df)
```

'+ 0'은 1로 구성된 상수항을 넣지 말라는 뜻이다. 만약 이 부분이 없으면 patsy는 자동으로 가장 앞에 Intercept란 이름의 상수항 데이터를 추가한다.
<br>
진짜 맨 앞에 Intercept라는 모든 열이 1인 상수항이 생긴다. <br>
이것을 `상수항 augmentation`이라고 부른다. 왜 필요한지는 회귀분석에서 나온다. <br>
회귀분석에서는 원래 이게 반드시 있어야 함. 굳이 빼고 싶어서 `+0` 혹은 `-1`이라고 하는 것. 


참고로 빼는 값이 카테고리 값은 경우에는 dummy 변수화를 해서 빼준다(one-hot).

```python
dmatrix("x1 + x2 + x3", data=df)
```

 1도 + 0과 같은 뜻이다.

```python
dmatrix("x1 + x2 + x3 - 1", data=df)
```

<br>
파이썬 코딩하듯이 그냥 그대로 문자열 안에 넣으면 됨. ***파이썬이 해석할 수 있는 글자는 다 가능하다.***

numpy 함수 뿐 아니라 사용자 정의 함수도 사용할 수 있다.

```python
def ten_times(x):
    return 10 * x

dmatrix("ten_times(x1)", df)
```

보통은 여러 열이 있을 때, 그 사이에서 합치고 곱하고 하면서 새로운 열들을 만들어낸다. 곱한값이 새로운 feature가 되고 그런 상황 많음. 
`x1, x2, x1*x2` 


기존의 데이터를 조합 연산하여 새로운 데이터를 만드는 것도 가능하다. 특히 두 변수를 곱해서 만들어지는 새로운 변수를 **상호작용(interaction)**이라고 한다. 상호작용은 : 기호를 사용하여 만들어진다.

이 곱하기하는 것을 특별히 상호작용이라고 부른다. 아주 자주 쓰기 때문에 interaction이라는 따로 이름까지 붙어 있는 것. 

회귀분석 할때도, 상호분석항들이 어떤 영향을 미치는지 살펴봐야 한다. 

```python
dmatrix("x1 + x2 + x1:x2 + 0", df)
```

두 변수와 상호작용을 한꺼번에 표시할 때는 *를 사용한다. 따라서 다음 수식은 바로 전의 수식과 동일하다.

즉, `x1/x2/x1:x2` 세개를 한번에 뽑고 싶을때, 간단하게 `x1*x2`로 표현한다. 

```python
dmatrix("x1 * x2 + 0", df)
```

상호작용을 제외한 경우에는 I()라는 연산자를 사용하여 연산과정을 명시해야 한다. 예를 들어 두 변수 x1과 x2를 더하여 새로운 데이터를 만들고 싶다면 다음과 같이 한다.

지금 여기서 +는 기본적으로 and임(x1과 x2를 뽑겠다). 그런데 그게 아니라, 실제로 두개를 더한 값을 새롭게 계산해서 뽑고 싶은 경우 I를 붙여줘야 한다.
+뿐만이 아니라, 여러가지 연산이 가능하다. ***I안에 있는 연산기호는 실제 연산기호로 해석한다.***

```python
dmatrix("x1 + x2 + I(x1 + x2) + 0", df)
```

##### Scaling

특히 회귀분석에서, condition numbe라는 것의 영향을 미치게 된다. 단위가 달라서, 크기가 제각각인게 영향을 미쳐서 조건수 라는 것이 나빠질 수 있다. <br>
기본적으로 평균이 0이 되고, 표준편차가 1이 되도록 단위를 바꿔준다. 

선형회귀분석을 할 때는 조건수(condition number)의 영향때문에 데이터의 평균을 0으로 표준편차를 1로 만드는 스케일링(scaling) 작업을 하는 것이 분석 결과의 품질을 높일 수 있다. patsy 패키지는 스케일링을 위한 함수도 제공한다.

center(): 평균을 0으로 스케일링

standardize(): 평균을 0으로하고 표준편차를 1로 스케일링

scale(): standardize() 과 같음


예를 들어 x1 데이터의 평균을 제거하는 변환은 다음과 같다(center는 각 데이터에서 평균을 싹다 빼주는 것. 그러면 당연히 평균이 0이 되겠지).

```python
dm = dmatrix("center(x1) + 0", df)
dm
```

이 변환 연산은 다음과 같이 x1 데이터에서 x1의 평균을 빼는 것과 같다.

```python
df.x1 - np.mean(df.x1)
```

스케일링을 하면, 평균이나 표준편차를 계산해 쓰는데, 그 값들은 이 안에 저장이 되어 있다. <br>
**왜 저장을 할까?**

***x -> scale() -> predict/training을 통한 model 생성 -> y***

모델을 만든 이유가 뭐야? 학습용 데이터의 y를 예측하려는 것이 아니라, y를 모르는 실제 데이터를 모형에 집어넣어서 y를 뽑고 싶은 것이 목적. <br>
이 모델이 스케일링 된 상태로 학습됬으니깐, 이 모델이 기대하는 것은 **"꽃잎의 길이에서 10을 뺀 값을 입력으로 받는 것을 기대중"**<br>
그런데, 실제 데이터로 예측을 하면서, 실제 데이터셋은 평균이 5라고 5를 뺀 값을 모델에 집어넣으면 돼?<br>
그러면 모델에 들어가는 값을 원래 모델에 들어갔어야 했던 값보다 5가 다 크게 들어가는 것. <br>

고로 테스트 데이터를 넣을 때도, 우리가 처음에 학습할때 썻던 스케일링을 해야함. <br>
당연히 테스트 데이터 넣는데 걔를 걔 나름대로 따로 스케일링 하면 절대 안되지. <br>
스케일링하는 평균이나 표준편차는 학습에서 썻던 값을 그대로 써야함. <br>
그것을 여기서 자동으로 저장을 해준다는 것. <br>
물론 내가 따로 빼서 하는 것은 아니고, 추후에 나오겠지만 그것또한 자동으로 해준다.
